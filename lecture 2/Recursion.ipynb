```python

def add(n):
    if n == 0:
        return 0
    else:
        return n + add(n - 1)


nums = list(range(10))


def arr_sum(arr, n):
    if n <= 0:
        return 0
    else:
        return arr[n - 1] + arr_sum(arr, n - 1)


print(arr_sum(nums, len(nums)))


def itr_reverse(arr):
    n = len(arr)
    for i in range(int(n / 2)):
        arr[i], arr[n - 1 - i] = arr[n - 1 - i], arr[i]
    return arr


itr_reverse(nums)


def rec_reverse(arr, i, n):
    if i > len(arr) / 2:
        return arr
    arr[i], arr[n - 1 - i] = arr[n - 1 - i], arr[i]
    return rec_reverse(arr, i + 1, n)


n = len(nums)
rec_reverse(nums, 0, n)


def func(x):
    return 3 * x + 1


def rec_func(func, x, n):
    if n <= 1:
        return func(x)
    else:
        return rec_func(func, func(x), n - 1)


print(rec_func(func, 5, 3))


class TreeNode:
    def __init__(self, val):
        self.left = None
        self.right = None
        self.val = val


tree = TreeNode(1)
tree.left = TreeNode(2)
tree.right = TreeNode(3)

tree.left.left = TreeNode(4)
tree.left.right = TreeNode(5)

tree.right.left = TreeNode(6)
tree.right.right = TreeNode(7)


def pre_order(tree):
    if tree:
        print(tree.val)
        pre_order(tree.left)
        pre_order(tree.right)


def in_order(tree):
    if tree:
        in_order(tree.left)
        print(tree.val)
        in_order(tree.right)


def post_order(tree):
    if tree:
        in_order(tree.left)
        in_order(tree.right)
        print(tree.val)


pre_order(tree)
print("======================" * 5)
in_order(tree)
print("======================" * 5)
post_order(tree)

```
